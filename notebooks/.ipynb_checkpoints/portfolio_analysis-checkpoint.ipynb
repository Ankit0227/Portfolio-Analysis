```python
import yfinance as yf
import pandas as pd
import time

stocks = ['AAPL', 'MSFT', 'GOOGL', 'TSLA', 'JPM', 'PFE', 'WMT', 'VTI']

start_date = '2019-01-01'
end_date = '2025-12-31'

print(f"Attempting to download {len(stocks)} stocks...\n")

all_data = {}
successful = []
failed = []

for stock in stocks:
    print(f"Downloading {stock}...")
    try:
        data = yf.download(stock, 
                          start=start_date, 
                          end=end_date,
                          progress=False)
        
        # Check if data is empty
        if len(data) == 0:
            print(f"⚠️  {stock}: Downloaded but got NO DATA")
            failed.append(stock)
        else:
            all_data[stock] = data
            successful.append(stock)
            print(f"✅ {stock}: Got {len(data)} trading days")
        
        time.sleep(2)
    
    except Exception as e:
        print(f"❌ {stock}: Error - {e}")
        failed.append(stock)

print(f"\n" + "=" * 80)
print(f"DOWNLOAD SUMMARY")
print("=" * 80)
print(f"Successful: {len(successful)} stocks")
print(f"  {successful}")
print(f"\nFailed: {len(failed)} stocks")
print(f"  {failed}")
print("=" * 80)

# Only combine if we have data
if len(all_data) > 0:
    combined_data = pd.concat(all_data, axis=1)
    print(f"\n✅ Combined data shape: {combined_data.shape}")
    print(f"Trading days: {combined_data.shape[0]}")
    print(f"Data columns: {combined_data.shape[1]}")
    
    combined_data.to_csv('data/raw/portfolio_prices.csv')
    print(f"\n✅ Saved to CSV!")
else:
    print(f"\n❌ NO DATA DOWNLOADED! Check your internet connection.")
```

```python
import pandas as pd
import os

print("STEP 1: CLEANING RAW DATA\n")

# Ensure folder exists
os.makedirs('data/processed', exist_ok=True)

# Load with MultiIndex columns
df = pd.read_csv('data/raw/portfolio_prices.csv', header=[0, 1])

# Remove junk rows
df = df[
    (df[('Unnamed: 0_level_0', 'Price')] != 'Date') &
    (df[('Unnamed: 0_level_0', 'Price')] != 'Ticker')
]

# Extract and clean date column
date_col = pd.to_datetime(df[('Unnamed: 0_level_0', 'Price')], errors='coerce')

# Keep only valid dates
df = df[date_col.notna()]
date_col = date_col[date_col.notna()]

# Drop original column
df = df.drop(columns=[('Unnamed: 0_level_0', 'Price')])

# Set index
df.index = date_col

# Extract Close prices
close_prices = df.xs('Close', level=1, axis=1)

# Drop missing
close_prices = close_prices.dropna()

# Save
close_prices.to_csv('data/processed/portfolio_prices_clean.csv')

print("✅ Data cleaned!")
print(f"Shape: {close_prices.shape}")
print(f"Date range: {close_prices.index[0]} to {close_prices.index[-1]}")
```

```python
import pandas as pd
import numpy as np

print("\nSTEP 2: CALCULATING ALL METRICS\n")

prices = pd.read_csv('data/processed/portfolio_prices_clean.csv', index_col=0, parse_dates=True)

daily_returns = prices.pct_change().dropna()

annual_return = daily_returns.mean() * 252
annual_volatility = daily_returns.std() * np.sqrt(252)

risk_free_rate = 0.04
sharpe_ratio = (annual_return - risk_free_rate) / annual_volatility

# Drawdown
cumulative = (1 + daily_returns).cumprod()
running_max = cumulative.cummax()
drawdown = (cumulative - running_max) / running_max
max_drawdown = drawdown.min()

# Beta (optimized)
market_returns = daily_returns['VTI']
market_variance = market_returns.var()

betas = {}
for stock in daily_returns.columns:
    if stock == 'VTI':
        betas[stock] = 1.0
    else:
        cov = daily_returns[[stock, 'VTI']].cov().iloc[0, 1]
        betas[stock] = cov / market_variance

correlation = daily_returns.corr()

summary = pd.DataFrame({
    'Annual Return': annual_return,
    'Annual Volatility': annual_volatility,
    'Sharpe Ratio': sharpe_ratio,
    'Max Drawdown': max_drawdown,
    'Beta': pd.Series(betas)
})

print(summary.round(4))

summary.to_csv('data/processed/metrics_summary.csv')
daily_returns.to_csv('data/processed/daily_returns.csv')
correlation.to_csv('data/processed/correlation_matrix.csv')

print("\n✅ STEP 2 COMPLETE")
```
